```python
file = "non_sensitive_data.txt"
with open(file, encoding="utf8") as f:
    line = f.read()
    rows = line.split("\n")
    
    news_id = []
    category = []
    text = []
    label = []
    for row in rows:
        row = row.split("_!_")
        news_id.append(row[0])
        category.append(row[2])
        text.append(row[3])
        label.append("non-sensitive")
        print(row)
        
        
    f.close()
    

    
```

```python
import pandas as pd

data = pd.DataFrame()
data["text"] = text
data["category"] = category
data["id"] = news_id
data["label"] = label

data.to_csv("non_sensitive_data.csv", encoding='utf_8_sig', index=False)
display(data)
print("Done!")
```
